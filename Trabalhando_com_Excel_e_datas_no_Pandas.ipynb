##**Tratando valores faltantes**

```python
#consultando linhas com valores faltantes
df.isnull().sum()
```

```python
#substituindo os valores nulos pela média
df["Vendas"].fillna(df["Vendas"].mean(), inplace=True)
```

```python
#verificando se os valores nulos foram substituídos
df.isnull().sum()
```

```python
#calculando a média
df["Vendas"].mean()
```

```python
#substituindo valores por 0
df["Vendas"].fillna(0, inplace=True)
```

```python
df.shape
```

```python
#apagando linhas com valores nulos
df.dropna(subset=["Vendas"], inplace=True)
```

##**Testando a remoção das linhas nulas**

```python
#lendo novamente os arquivos
df1 = pd.read_excel("Aracaju.xlsx")
df2 = pd.read_excel("Fortaleza.xlsx")
df3 = pd.read_excel("Natal.xlsx")
df4 = pd.read_excel("Recife.xlsx")
df5 = pd.read_excel("Salvador.xlsx")
```

```python
#juntando (concatenando) todos os arquivos
df = pd.concat([df1,df2,df3,df4,df5])
```

```python
#verificando
df.shape
```

*Criando colunas novas*

```python
#criando a coluna da receita
df["Receita"]= df["Vendas"].mul(df["Qtde"])
```

```python
#validando
df.head()
```

```python
#descobrindo a quantidade
df["Receita/Vendas"] = df["Receita"]/ df["Vendas"]
```

```python
df.head()
```

```python
#retornando a maior receita
df["Receita"].max()
```

```python
#retornando a menor receita
df["Receita"].min()
```

```python
#trazendo as maiores receitas nlargest
df.nlargest(5, "Receita")
```

```python
#trazendo as menores receitas nsmallest
df.nsmallest (5, "Receita")
```

```python
#Agrupando por cidade e somando as receitas
df.groupby("Cidade")["Receita"].sum()
```

```python
#ordenando o conjunto de dados
df.sort_values("Receita", ascending=False).head(10)
```

#**TRABALHANDO COM DATAS**

```python
#transformando a coluna data em tipo inteiro
df["Data"] = df["Data"].astype("int64")
```

```python
#verificando o tipo de dado de cada coluna, novamente.
df.dtypes
```

```python
#transformando a coluna data em data novamente
df["Data"] = pd.to_datetime(df["Data"])
```

```python
#verificando a alteração
df.dtypes
```

```python
#agrupando por ano
df.groupby(df["Data"].dt.year)["Receita"].sum()
```

```python
#criando nova coluna com ano
df["Ano_Venda"] = df["Data"].dt.year
```

```python
#verificando amostra
df.sample(5)
```

```python
#Extraindo o mes e o dia
df["mês_venda"], df["dia_venda"] = (df["Data"].dt.month, df["Data"].dt.day)
```

```python
#verificando amostra
df.sample(5)
```

```python
#retornando a data mais antiga
df["Data"].min()
```

```python
#calculando a diferença de dias da data mínima
df["diferença_dias"] = df["Data"] - df["Data"].min()
df.sample(3)
```

```python
#filtrando vendas de 2019 do mês de Março
vendas_marco_19 = df.loc[(df["Data"].dt.year == 2019) & (df["Data"].dt.month == 3)]
vendas_marco_19
```
