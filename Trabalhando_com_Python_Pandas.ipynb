**Aula_1 com Python e Pandas**

```python
#importando a biblioteca pandas
import pandas as pd
```

```python
#importando a base de dados CSV para análise
df = pd.read_csv("/content/drive/MyDrive/datasets/Gapminder.csv",error_bad_lines=False, sep=";")
```

```python
#Visualisando as 5 primeiras linhas
df.head()
```

```python
#Renomeando as colunas
df = df.rename(columns={"country":"pais", "continent":"continente", "year":"ano", "lifeExp":"Expectativa_Vida","pop":"Pop_Total", "gdpPercap":"PIB"})
```

```python
#Visualizando as 15 primeiras linhas
df.head(15)
```

```python
#visualizando o total de linhas e colunas
df.shape
```

```python
#verificando quais os nomes das colunas
df.columns
```

```python
#verificando os tipos de dados
df.dtypes
```

```python
#retornando as ultimas linhas de dados
df.tail()
```

```python
#retornando as 15 ultimas linhas de dados
df.tail(15)
```

```python
#retornando analise estatística dos dados
df.describe()
```

```python
#realizando filtros = retornando a lista de valores da coluna continente
df["continente"].unique()
```

```python
#retornando apenas os valores do continente Europa e retornando os 4 primeiros valores
Europa = df.loc[df["continente"] == "Europe"]
Europa.head(4)
```

```python
#Verificando se os dados retornados da variavel "Europa" são realmente apenas do continente europeu:
Europa["continente"].unique()
```

```python
#agrupamento de dados: agrupando a quantidade de países por continente
df.groupby("continente")["pais"].nunique()
```

```python
#agrupanmento de dados: verificando a expectativa de vida média para cada ano
df.groupby("ano")["Expectativa_Vida"].mean()
```

```python
#calculando média do PIB
df["PIB"].mean()
```

```python
#calculando a soma do PIB
df["PIB"].sum()
```
